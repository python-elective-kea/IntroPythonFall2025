```python
data = {
    "Department": ["HR", "IT", "Finance", "IT", "HR"],
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "Salary": [70000, 80000, 90000, 85000, 72000],
    "YearsExperience": [2, 5, 7, 4, 3]
}
df = pd.DataFrame(data)

df.groupby('Department')['Salary'].mean()

df.groupby('Department')['YearsExperience'].sum()
df.groupby('Department')['YearsExperience'].sum().sort_values(ascending=False)
```
