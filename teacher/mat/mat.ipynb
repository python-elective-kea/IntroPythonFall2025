```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values
x = np.linspace(-6, 10, 100)

# Define the coefficients of the polynomial y = ax^2 + bx + c
coefficients = [1, -2, 3]  # a=1, b=-2, c=3

# Calculate y values
y = np.polyval(coefficients, x)

# Plot the polynomial
plt.plot(x, y, label="y = x^2 - 2x + 3")
plt.xlabel("x")
plt.ylabel("y")
plt.title("2nd Degree Polynomial Plot")
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define x values (including negative values)
x = range(-6, 7)  # Values from -6 to 6

# Define the coefficients of the polynomial y = ax^2 + bx + c
a, b, c = 1, -2, 3  # a=1, b=-2, c=3

# Calculate y values using plain Python
y = [(a * i**2 + b * i + c) for i in x]

# Plot the polynomial
plt.plot(x, y, label="y = x^2 - 2x + 3", marker='o')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Add x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # Add y-axis
plt.xlabel("x")
plt.ylabel("y")
plt.title("2nd Degree Polynomial Plot with Negative x Values")
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

# Create a figure with a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot data on each subplot
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Sine Wave')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Cosine Wave')

axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Tangent Wave')

axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Exponential Function')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the figure
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definer en firkant (fire hjørner som vektorer)
square = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1],
    [0, 0]  # Luk firkanten
])

# Plot original firkant
plt.plot(square[:, 0], square[:, 1], label='Original', linestyle='--')

# Transformationer
# Skalering
scale_factor = 2
scaled_square = square * scale_factor

# Rotation (90 grader med urets retning)
theta = np.radians(90)
rotation_matrix = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta), np.cos(theta)]
])
rotated_square = np.dot(square, rotation_matrix)

# Translation (flytning)
translation_vector = np.array([2, 2])
translated_square = square + translation_vector

# Plot de transformerede figurer
plt.plot(scaled_square[:, 0], scaled_square[:, 1], label='Skaleret')
plt.plot(rotated_square[:, 0], rotated_square[:, 1], label='Roteret')
plt.plot(translated_square[:, 0], translated_square[:, 1], label='Flyttet')

# Konfiguration af plot
plt.axhline(0, color='gray', linewidth=0.5, linestyle='--')
plt.axvline(0, color='gray', linewidth=0.5, linestyle='--')
plt.legend()
plt.title("Manipulation af en firkant")
plt.xlabel("x-akse")
plt.ylabel("y-akse")
plt.axis('equal')
#plt.grid(True)

# Vis plot
plt.show()
```
