### figure
* En Figure er hele det visuelle objekt, se det som et lærred.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4]
y = [1,2,3,4]


plt.title('Anatomy of a Figure')
plt.plot(x, y,  label='y=x*a+b')

plt.xticks(range(0,5))
plt.yticks(range(0,5))

# minor ticks
plt.minorticks_on()
minor_ticks = np.arange(0, 4.25, 0.25)
plt.gca().xaxis.set_ticks(minor_ticks, minor=True)

plt.gca().yaxis.set_ticks(minor_ticks, minor=True)
plt.gca().xaxis.set_ticklabels([f"{tick}" for tick in minor_ticks], minor=True)

plt.gca().yaxis.set_ticks(minor_ticks, minor=True)
plt.tick_params(axis='x', which='minor', labelsize=8, labelcolor='blue')

# Label
plt.xlabel('X axis label')
plt.ylabel('Y axis label')

# Legend
plt.legend()

```

### Forskel på pyplot-style og OO-style

https://matplotlib.org/3.5.0/tutorials/introductory/usage.html#the-object-oriented-interface-and-the-pyplot-interface

#### OO-Style

```python
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

#### Pyplot-style

```python
x = np.linspace(0, 2, 100)  # Sample data.

plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt

# Læs lydfil
rate, data = wavfile.read('src/Bongo_sound.wav')

# Skaler lydniveauet
scaled_data = data * 10

# Plot original og skaleret lyd
time = np.linspace(0., len(data) / rate, len(data))
plt.plot(time, data, label="Original")
plt.plot(time, scaled_data, label="Skaleret")
plt.legend()
plt.show()

# Gem den skalerede lyd
wavfile.write('output.wav', rate, scaled_data.astype(np.int16))
```
